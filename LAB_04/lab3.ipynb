## L thuyt
Thut ton di truyn (Genetic Algorithm - GA) l mt phng php tm kim da trn c ch tin ha t nhin, ly cm hng t thuyt tin ha ca Darwin. GA c s dng  tm li gii gn ti u cho cc bi ton ti u ha phc tp.

### Cc bc chnh ca thut ton:
1. **Khi to qun th**: To ngu nhin mt tp hp cc c th (gii php tim nng).
2. **nh gi  thch nghi (Fitness Evaluation)**: Tnh gi tr hm mc tiu cho mi c th.
3. **La chn (Selection)**: Chn cc c th tt nht  lai ghp (thng s dng cc phng php nh Tournament Selection hoc Roulette Wheel).
4. **Lai ghp (Crossover)**: Kt hp hai c th  to ra c th mi.
5. **t bin (Mutation)**: Thay i ngu nhin mt s gen  duy tr tnh a dng.
6. **Lp li**: Quay li bc 2 cho n khi t iu kin dng (s th h hoc gi tr fitness mong mun).

## V d 1: Ti u ha hm mt bin

### Bi ton

$$

$$

# M gi
```plaintext
To qun th mi rng
Lp cho n khi qun th mi  N c th:
    Chn cha m 1 bng Tournament Selection:
        - Chn ngu nhin tournament_size c th trong qun th
        - Chn c th c fitness cao nht lm cha m 1
    Chn cha m 2 bng Tournament Selection tng t
    
    Thc hin lai ghp (crossover) gia cha m 1 v cha m 2 vi xc sut crossover_rate:
        Nu crossover thnh cng:
            To con mi c gi tr trung bnh ca cha m
        Ngc li:
            Con mi l cha m 1
            
    Thc hin t bin cho con mi vi xc sut mutation_rate:
        Nu t bin:
            Gn con mi gi tr ngu nhin trong [min_val, max_val]
            
    Thm con mi vo qun th mi
    
Cp nht qun th hin ti = qun th mi
In ra c th tt nht v gi tr fitness tng ng ca th h


````python
import numpy as np
import random
import matplotlib.pyplot as plt

# Hm mc tiu
def fitness_function(x):
    return -(x**2) + 10*x + 50

# Khi to qun th
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# La chn (Tournament Selection)
def select_parents(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghp
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# t bin
def mutate(individual, mutation_rate=0.1, min_val=0, max_val=10):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thut ton di truyn
def genetic_algorithm_example1(pop_size=50, generations=100, min_val=0, max_val=10):
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents(population, fitness_values)
            parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function(x) for x in population])]
        print(f'Th h {generation + 1}: x = {best_individual:.4f}, f(x) = {best_fitness:.4f}')
    
    best_fitness = max([fitness_function(x) for x in population])
    best_individual = population[np.argmax([fitness_function(x) for x in population])]
    
    # Trc quan ha
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr f(x) tt nht')
    plt.title('S thay i ca f(x) qua cc th h')
    plt.show()
    
    return best_individual, best_fitness

# Chy thut ton
best_x, best_f = genetic_algorithm_example1()
print(f'\nKt qu cui cng: x = {best_x:.4f}, f(x) = {best_f:.4f}')
````

## V d 2: Ti u ha hm hai bin

### Bi ton

### M gi 

```plaintext
To qun th mi rng
Lp cho n khi qun th mi  N c th:
    Chn cha m 1 bng Tournament Selection trn qun th hin ti
    Chn cha m 2 bng Tournament Selection trn qun th hin ti
    Thc hin lai ghp (crossover) gia cha m 1 v cha m 2 vi xc sut crossover_rate:
        Nu crossover thnh cng:
            To con mi bng trung bnh gi tr (x, y) ca cha m
        Ngc li:
            Con mi l cha m 1
    Thc hin t bin cho con mi vi xc sut mutation_rate tng bin:
        Nu t bin:
            Gn gi tr ngu nhin trong [min_val, max_val] cho bin tng ng
    Thm con mi vo qun th mi

Cp nht qun th hin ti = qun th mi
In kt qu tt nht th h

````python
# Hm mc tiu
def fitness_function_2d(x, y):
    return -(x**2 + y**2)  # S dng du m v GA tm max

# Khi to qun th
def initialize_population_2d(pop_size, min_val, max_val):
    return [(random.uniform(min_val, max_val), random.uniform(min_val, max_val)) for _ in range(pop_size)]

# La chn
def select_parents_2d(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghp
def crossover_2d(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return ((parent1[0] + parent2[0]) / 2, (parent1[1] + parent2[1]) / 2)
    return parent1

# t bin
def mutate_2d(individual, mutation_rate=0.1, min_val=-5, max_val=5):
    x, y = individual
    if random.random() < mutation_rate:
        x = random.uniform(min_val, max_val)
    if random.random() < mutation_rate:
        y = random.uniform(min_val, max_val)
    return (x, y)

# Thut ton di truyn
def genetic_algorithm_example2(pop_size=50, generations=100, min_val=-5, max_val=5):
    population = initialize_population_2d(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function_2d(x, y) for x, y in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(-best_fitness)  # Chuyn v gi tr thc
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents_2d(population, fitness_values)
            parent2 = select_parents_2d(population, fitness_values)
            child = crossover_2d(parent1, parent2)
            child = mutate_2d(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
        print(f'Th h {generation + 1}: (x, y) = ({best_individual[0]:.4f}, {best_individual[1]:.4f}), g(x, y) = {-best_fitness:.4f}')
    
    best_fitness = max([fitness_function_2d(x, y) for x, y in population])
    best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
    
    # Trc quan ha
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr g(x, y) tt nht')
    plt.title('S thay i ca g(x, y) qua cc th h')
    plt.show()
    
    return best_individual, -best_fitness

# Chy thut ton
best_xy, best_g = genetic_algorithm_example2()
print(f'\nKt qu cui cng: (x, y) = ({best_xy[0]:.4f}, {best_xy[1]:.4f}), g(x, y) = {best_g:.4f}')
````

# Bi tp 1
# Thay i hm mc tiu**: Th p dng thut ton cho hm $h(x) = sin(x) + cos(x)$ trong khong $x \in [0, 2\pi]$. So snh kt qu vi nghim chnh xc.

## Ti u hm $h(x) = \sin(x) + \cos(x)$ bng Thut Ton Di Truyn (Genetic Algorithm)

## Mc tiu
- Ti u hm:
  $$
  h(x) = \sin(x) + \cos(x)
  $$
  trong min:
  $$
  x \in [0, 2\pi]
  $$
- Tm gi tr **x** sao cho **h(x)** t gi tr ln nht.

---

## Phn tch nghim chnh xc

Ta c:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

##  tng thut ton di truyn (GA)

1. **M ha**: Biu din gi tr \( x \) bng chui nh phn di 16 bit.  
2. **Gii m**: Chuyn t chui nh phn sang gi tr thc trong khong \([0, 2\pi]\).  
3. **Fitness**: S dng chnh gi tr \( h(x) \) lm  thch nghi.  
4. **Chn lc**: Roulette Wheel Selection.  
5. **Lai ghp**: Lai 1 im vi xc sut cao.  
6. **t bin**: Lt bit vi xc sut thp.

---


````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# code vo  y nhe


# --- 1. Ci t cc tham s cho thut ton ---
# Min gi tr ca x
BOUNDS = [0, 2 * np.pi]
# S lng bit  m ha mt c th (gi tr x)
NUM_BITS = 16
# Kch thc qun th
POP_SIZE = 50
# T l lai ghp
CROSSOVER_RATE = 0.8
# T l t bin
MUTATION_RATE = 0.01
# S th h
NUM_GENERATIONS = 100

# --- 2. nh ngha cc hm ct li ---

def objective_function(x):
    #Hm mc tiu h(x) cn ti u.#
    return np.sin(x) + np.cos(x)

def decode(bitstring, bounds, n_bits):
    #Gii m chui bit thnh gi tr thc x trong min [min, max].#
    # Chuyn chui bit thnh s nguyn
    integer_val = int("".join(map(str, bitstring)), 2)
    
    # Tnh gi tr ln nht m s nguyn c th t c
    max_int = 2**n_bits - 1
    
    # nh x gi tr nguyn sang min gi tr thc
    span = bounds[1] - bounds[0]
    decoded_val = bounds[0] + (integer_val / max_int) * span
    return decoded_val

def selection(pop, fitnesses): # Chn lc c th bng phng php Roulette Wheel.
    # Dch chuyn fitness  tt c u khng m
    min_fitness = np.min(fitnesses)
    shifted_fitnesses = fitnesses - min_fitness
    
    total_fitness = np.sum(shifted_fitnesses)
    
    # Nu tt c cc fitness bng nhau, tr v la chn ngu nhin
    if total_fitness == 0:
        return pop[np.random.randint(0, len(pop), size=len(pop))]

    # Tnh xc sut chn lc
    probabilities = shifted_fitnesses / total_fitness
    
    # Chn lc cc c th mi
    indices = np.random.choice(len(pop), size=len(pop), p=probabilities)
    return pop[indices]

# Lai ghp 1 im
def crossover(p1, p2, rate):
    c1, c2 = p1.copy(), p2.copy()
    if np.random.rand() < rate:
        # Chn ngu nhin 1 im  lai ghp
        pt = np.random.randint(1, len(p1) - 1)
        # Thc hin lai ghp
        c1 = np.concatenate((p1[:pt], p2[pt:]))
        c2 = np.concatenate((p2[:pt], p1[pt:]))
    return c1, c2

def mutation(bitstring, rate):
    #Thc hin t bin bng cch lt bit.#
    for i in range(len(bitstring)):
        if np.random.rand() < rate:
            # Lt bit (0 -> 1, 1 -> 0)
            bitstring[i] = 1 - bitstring[i]
    return bitstring

# --- 3. Chy thut ton di truyn ---

# Khi to qun th ban u
population = np.random.randint(0, 2, size=(POP_SIZE, NUM_BITS))

# Lu li c th tt nht qua cc th h
best_x, best_h = 0, -np.inf
best_fitness_history = []

print("Bt u qu trnh ti u...")
for gen in range(NUM_GENERATIONS):
    # 1. Gii m v nh gi (tnh fitness)
    decoded_pop = np.array([decode(p, BOUNDS, NUM_BITS) for p in population])
    fitnesses = objective_function(decoded_pop)

    # 2. Tm c th tt nht trong th h hin ti v ton cc
    current_best_idx = np.argmax(fitnesses)
    if fitnesses[current_best_idx] > best_h:
        best_h = fitnesses[current_best_idx]
        best_x = decoded_pop[current_best_idx]
        
    best_fitness_history.append(best_h)
    
    if (gen + 1) % 10 == 0:
        print(f"> Th h {gen+1:3d}, x tt nht = {best_x:.4f}, h(x) = {best_h:.4f}")

    # 3. Chn lc
    mating_pool = selection(population, fitnesses)
    
    # 4. To th h mi
    new_population = []
    for i in range(0, POP_SIZE, 2):
        p1, p2 = mating_pool[i], mating_pool[i+1]
        
        # 5. Lai ghp
        c1, c2 = crossover(p1, p2, CROSSOVER_RATE)
        
        # 6. t bin
        c1 = mutation(c1, MUTATION_RATE)
        c2 = mutation(c2, MUTATION_RATE)
        
        new_population.append(c1)
        new_population.append(c2)
        
    population = np.array(new_population)

# --- 4. In kt qu v so snh ---
print("\n--- KT QU CUI CNG ---")

# Nghim chnh xc
x_exact = np.pi / 4
h_exact = np.sqrt(2)

print(f"Nghim tm thy bi GA: ")
print(f"  x = {best_x:.6f}")
print(f"  h(x) = {best_h:.6f}")
print("-" * 20)
print(f"Nghim chnh xc (l thuyt):")
print(f"  x = {x_exact:.6f} (/4)")
print(f"  h(x) = {h_exact:.6f} (2)")
print("-" * 20)
print(f"Sai s tuyt i cho x: {abs(best_x - x_exact):.6f}")
print(f"Sai s tuyt i cho h(x): {abs(best_h - h_exact):.6f}")

# --- 5. Trc quan ha ---
plt.figure(figsize=(10, 6))
plt.plot(best_fitness_history, label='Gi tr h(x) tt nht mi th h')
plt.axhline(y=h_exact, color='r', linestyle='--', label=f'Gi tr Max chnh xc (2  {h_exact:.4f})')
plt.xlabel("Th h")
plt.ylabel("Gi tr h(x) (Fitness)")
plt.title("S hi t ca Thut ton Di truyn")
plt.legend()
plt.grid(True)
plt.show()
````

# Bi tp 2: Ti u hm $h(x) = \sin(x) + \cos(x)$ bng Thut Ton Di Truyn (Genetic Algorithm)

---

## Mc tiu

Ti u hm:
$$
h(x) = \sin(x) + \cos(x)

h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xy ra khi:
$$

##   tng thut ton di truyn (GA)

| Thnh phn        | Chi tit                                                                 |
|------------------|-------------------------------------------------------------------------|
| **M ha**        | Dng chui nh phn 16 bit  biu din gi tr $x \in [0, 2\pi]$         |
| **Gii m**       | nh x nh phn sang gi tr thc: $x = 0 + \dfrac{\text{decimal}}{2^{16}-1} \cdot 2\pi$ |

````python


# --- 1. Ci t cc tham s cho thut ton ---
# Min gi tr ca x
BOUNDS = [0, 2 * np.pi]
# S lng bit  m ha mt c th (gi tr x)
NUM_BITS = 16
# Kch thc qun th
POP_SIZE = 50
# T l lai ghp
CROSSOVER_RATE = 0.8
# T l t bin (THEO YU CU BI TP)
MUTATION_RATE = 0.05
# S th h
NUM_GENERATIONS = 100

# --- 2. nh ngha cc hm ct li ---

def objective_function(x):
    #Hm mc tiu h(x) cn ti u.#
    return np.sin(x) + np.cos(x)

def decode(bitstring, bounds, n_bits):
    #Gii m chui bit thnh gi tr thc x.#
    # Chuyn chui bit thnh s nguyn
    integer_val = int("".join(map(str, bitstring)), 2)
    # Gi tr ln nht m s nguyn c th t c
    max_int = 2**n_bits - 1
    # nh x gi tr nguyn sang min gi tr thc
    span = bounds[1] - bounds[0]
    decoded_val = bounds[0] + (integer_val / max_int) * span
    return decoded_val

def selection(pop, fitnesses):
    #Chn lc c th bng phng php Roulette Wheel.#
    # Dch chuyn fitness  tt c u khng m
    min_fitness = np.min(fitnesses)
    shifted_fitnesses = fitnesses - min_fitness
    
    total_fitness = np.sum(shifted_fitnesses)
    
    # Nu tt c cc fitness bng nhau, tr v la chn ngu nhin
    if total_fitness == 0:
        return pop[np.random.randint(0, len(pop), size=len(pop))]

    probabilities = shifted_fitnesses / total_fitness
    indices = np.random.choice(len(pop), size=len(pop), p=probabilities)
    return pop[indices]

def crossover(p1, p2, rate):
    #Thc hin lai ghp 1 im.#
    c1, c2 = p1.copy(), p2.copy()
    if np.random.rand() < rate:
        pt = np.random.randint(1, len(p1) - 1)
        c1 = np.concatenate((p1[:pt], p2[pt:]))
        c2 = np.concatenate((p2[:pt], p1[pt:]))
    return c1, c2

def mutation(bitstring, rate):
    #Thc hin t bin bng cch lt bit.#
    for i in range(len(bitstring)):
        if np.random.rand() < rate:
            bitstring[i] = 1 - bitstring[i]
    return bitstring

# --- 3. Chy thut ton di truyn ---

# Khi to qun th ban u
population = np.random.randint(0, 2, size=(POP_SIZE, NUM_BITS))

# Lu li c th tt nht qua cc th h
best_x, best_h = 0, -np.inf
best_fitness_history = []

print("Bt u qu trnh ti u vi Mutation Rate = 0.05...")
for gen in range(NUM_GENERATIONS):
    # 1. Gii m v nh gi
    decoded_pop = np.array([decode(p, BOUNDS, NUM_BITS) for p in population])
    fitnesses = objective_function(decoded_pop)

    # 2. Tm c th tt nht
    current_best_idx = np.argmax(fitnesses)
    if fitnesses[current_best_idx] > best_h:
        best_h = fitnesses[current_best_idx]
        best_x = decoded_pop[current_best_idx]
        
    best_fitness_history.append(best_h)
    
    if (gen + 1) % 10 == 0:
        print(f"> Th h {gen+1:3d}, x tt nht = {best_x:.4f}, h(x) = {best_h:.4f}")

    # 3. Chn lc
    mating_pool = selection(population, fitnesses)
    
    # 4. To th h mi (Lai ghp v t bin)
    new_population = []
    for i in range(0, POP_SIZE, 2):
        p1, p2 = mating_pool[i], mating_pool[i+1]
        c1, c2 = crossover(p1, p2, CROSSOVER_RATE)
        c1 = mutation(c1, MUTATION_RATE)
        c2 = mutation(c2, MUTATION_RATE)
        new_population.append(c1)
        new_population.append(c2)
        
    population = np.array(new_population)

# --- 4. In kt qu v so snh ---
print("\n--- KT QU CUI CNG ---")

# Nghim chnh xc
x_exact = np.pi / 4
h_exact = np.sqrt(2)

print(f"Nghim tm thy bi GA: ")
print(f"  x = {best_x:.6f}")
print(f"  h(x) = {best_h:.6f}")
print("-" * 20)
print(f"Nghim chnh xc (l thuyt):")
print(f"  x = {x_exact:.6f} (/4)")
print(f"  h(x) = {h_exact:.6f} (2)")
print("-" * 20)
print(f"Sai s tuyt i cho x: {abs(best_x - x_exact):.6f}")
print(f"Sai s tuyt i cho h(x): {abs(best_h - h_exact):.6f}")

# --- 5. Trc quan ha ---
plt.figure(figsize=(10, 6))
plt.plot(best_fitness_history, label=f'Gi tr h(x) tt nht (Mutation Rate = {MUTATION_RATE})')
plt.axhline(y=h_exact, color='r', linestyle='--', label=f'Gi tr Max chnh xc (2  {h_exact:.4f})')
plt.xlabel("Th h")
plt.ylabel("Gi tr h(x) (Fitness)")
plt.title("S hi t ca Thut ton Di truyn")
plt.legend()
plt.grid(True)
plt.show()
````

# Thay i tham s thut ton di truyn v phn tch tc  hi t

**Mc tiu:** Th thay i cc tham s chnh ca thut ton di truyn gm:

- Kch thc qun th: `pop_size`
- S th h: `generations`
- T l t bin: `mutation_rate`
- T l lai ghp: `crossover_rate`

v phn tch nh hng ca tng tham s n tc  hi t ca thut ton.

---

## Cc tham s v vai tr

| Tham s         |  ngha                                  | nh hng chnh                                                                                         |
|-----------------|-----------------------------------------|--------------------------------------------------------------------------------------------------------|
| `pop_size`      | S c th trong qun th                 | Qun th ln gip khm ph khng gian tm kim tt hn nhng tn nhiu thi gian tnh ton. Qun th nh hi t nhanh nhng d b kt cc tr cc b. |
| `generations`   | S vng lp ca thut ton               | S th h cng nhiu th thut ton c nhiu c hi ci thin nghim, tuy nhin thi gian chy cng tng. |
| `mutation_rate` | Xc sut t bin mi bit                | t bin cao gip tng a dng gene, trnh b kt trong cc tr cc b, nhng nu qu cao lm qu trnh tm kim nhiu lon. |
| `crossover_rate`| Xc sut lai ghp gia hai c th       | Lai ghp gip phi hp gen tt, tng kh nng tm nghim tt. Nu qu thp, qun th km a dng; qu cao c th ph hy cu trc tt ca c th. |

---

## Phn tch nh hng

- **`pop_size`**: Kch thc qun th ln gip tm kim ton cc tt hn nhng chm hn, kch thc nh lm thut ton hi t nhanh nhng d mc kt.
- **`generations`**: S th h ln tng kh nng ci tin kt qu nhng lm tng thi gian tnh ton.
- **`mutation_rate`**: T l t bin va phi gip duy tr a dng di truyn, trnh kt cc b; qu cao hoc qu thp u lm gim hiu qu.
- **`crossover_rate`**: T l lai ghp hp l gip khai thc tt thng tin di truyn, tng hiu qu hi t.

---




````python
# code vo y 

# --- HM CT LI CA GA (Gi nguyn t cc bi trc) ---
BOUNDS = [0, 2 * np.pi]
NUM_BITS = 16
H_EXACT = np.sqrt(2)

def objective_function(x):
    return np.sin(x) + np.cos(x)

def decode(bitstring, bounds, n_bits):
    integer_val = int("".join(map(str, bitstring)), 2)
    max_int = 2**n_bits - 1
    span = bounds[1] - bounds[0]
    return bounds[0] + (integer_val / max_int) * span

def selection(pop, fitnesses):
    min_fitness = np.min(fitnesses)
    shifted_fitnesses = fitnesses - min_fitness
    total_fitness = np.sum(shifted_fitnesses)
    if total_fitness == 0:
        return pop[np.random.randint(0, len(pop), size=len(pop))]
    probabilities = shifted_fitnesses / total_fitness
    indices = np.random.choice(len(pop), size=len(pop), p=probabilities)
    return pop[indices]

def crossover(p1, p2, rate):
    c1, c2 = p1.copy(), p2.copy()
    if np.random.rand() < rate:
        pt = np.random.randint(1, len(p1) - 1)
        c1 = np.concatenate((p1[:pt], p2[pt:]))
        c2 = np.concatenate((p2[:pt], p1[pt:]))
    return c1, c2

def mutation(bitstring, rate):
    for i in range(len(bitstring)):
        if np.random.rand() < rate:
            bitstring[i] = 1 - bitstring[i]
    return bitstring

# --- HM CHY TH NGHIM ---
def run_ga_experiment(pop_size, generations, crossover_rate, mutation_rate):
    
    # Chy mt lt thut ton GA vi cc tham s cho trc.
    # Tr v lch s gi tr fitness tt nht qua cc th h.
    population = np.random.randint(0, 2, size=(pop_size, NUM_BITS))
    best_fitness_history = []
    best_h = -np.inf

    for _ in range(generations):
        decoded_pop = np.array([decode(p, BOUNDS, NUM_BITS) for p in population])
        fitnesses = objective_function(decoded_pop)
        
        current_best_h = np.max(fitnesses)
        if current_best_h > best_h:
            best_h = current_best_h
        best_fitness_history.append(best_h)
        
        mating_pool = selection(population, fitnesses)
        
        new_population = []
        for i in range(0, pop_size, 2):
            p1, p2 = mating_pool[i], mating_pool[i+1]
            c1, c2 = crossover(p1, p2, crossover_rate)
            c1 = mutation(c1, mutation_rate)
            c2 = mutation(c2, mutation_rate)
            new_population.extend([c1, c2])
            
        population = np.array(new_population)
        
    return best_fitness_history

# --- HM  CHY V V BIU  ---
def analyze_parameter(param_name, param_values, base_params, n_runs=10):
    
    #Phn tch nh hng ca mt tham s bng cch chy nhiu ln v v biu .
    plt.figure(figsize=(12, 7))
    plt.title(f'Phn tch nh hng ca {param_name}', fontsize=16)
    
    for value in param_values:
        print(f"--- ang th nghim: {param_name} = {value} ---")
        current_params = base_params.copy()
        current_params[param_name] = value
        
        all_histories = []
        for i in range(n_runs):
           # print(f"  Run {i+1}/{n_runs}...")
            history = run_ga_experiment(**current_params)
            all_histories.append(history)
            
        # Ly trung bnh kt qu ca n_runs ln chy
        avg_history = np.mean(np.array(all_histories), axis=0)
        
        plt.plot(avg_history, label=f'{param_name} = {value}')

    plt.axhline(y=H_EXACT, color='r', linestyle='--', label=f'Gi tr Max chnh xc  {H_EXACT:.4f}')
    plt.xlabel("Th h")
    plt.ylabel("Gi tr h(x) trung bnh tt nht")
    plt.legend()
    plt.grid(True)
    plt.show()

# --- THC HIN CC TH NGHIM ---
if __name__ == "__main__":
    BASE_PARAMETERS = {
        'pop_size': 50,
        'generations': 50,
        'crossover_rate': 0.8,
        'mutation_rate': 0.01
    }

    # 1. Phn tch Kch thc qun th (pop_size)
    analyze_parameter(
        param_name='pop_size',
        param_values=[10, 50, 200],
        base_params=BASE_PARAMETERS
    )
    
    # 2. Phn tch T l t bin (mutation_rate)
    analyze_parameter(
        param_name='mutation_rate',
        param_values=[0.001, 0.01, 0.05],
        base_params=BASE_PARAMETERS
    )

    # 3. Phn tch T l lai ghp (crossover_rate)
    analyze_parameter(
        param_name='crossover_rate',
        param_values=[0.2, 0.8, 1.0],
        base_params=BASE_PARAMETERS
    )
````

# Bi tp 4
# Ci tin thut ton  Thm phng php la chn khc v so snh hiu qu

---

## Mc tiu

- Thm phng php la chn mi vo thut ton di truyn, v d nh **Roulette Wheel Selection** (La chn bnh xe con quay).
- So snh hiu qu ca phng php la chn mi vi phng php c (v d: Chn ngu nhin, Tournament Selection, v.v.).
- Phn tch nh hng ca phng php la chn n tc  hi t v cht lng nghim.

---

##  Cc phng php la chn ph bin

| Phng php                | M t                                                                                   | u im                              | Nhc im                          |
|----------------------------|-----------------------------------------------------------------------------------------|------------------------------------|-----------------------------------|
| Chn ngu nhin (Random)   | Chn c th ngu nhin trong qun th                                                   | n gin, nhanh                    | Khng u tin c th tt hn      |
| Roulette Wheel Selection    | Chn theo xc sut t l thun vi  thch nghi (fitness)                              | D dng, u tin c th tt        | D b nh hng bi c th qu mnh|
| Tournament Selection        | Chn ngu nhin mt s c th, chn c th tt nht trong nhm                          | Kim sot a dng tt              | Cn chn kch thc nhm ph hp  |

---

##   tng ci tin

1. Ci t thm phng php **Roulette Wheel Selection** trong thut ton di truyn.
2. Chy th vi cng b tham s trn cc phng php la chn khc nhau.
3. So snh kt qu v:
   - Tc  hi t (s th h cn  t nghim tt).
   - Gi tr nghim tt nht t c.
4. Phn tch u  nhc im tng phng php trong thc t.

---

##  Kt qu v phn tch

- Phng php **Roulette Wheel** thng gip tng tc  hi t nh u tin c th tt.
- Tuy nhin c th lm gim a dng gen nu c th u t chim u th qu ln, d ri vo cc tr cc b.
- Phng php **Tournament** kim sot tt s a dng nhng c th chm hn trong mt s trng hp.
- Phng php **Random** n gin nhng t hiu qu trong ti u nng cao.

---


````python
# code vo y nh
import numpy as np
import matplotlib.pyplot as plt
from functools import partial

# --- HM CT LI CA GA (Gi nguyn) ---
BOUNDS = [0, 2 * np.pi]
NUM_BITS = 16
H_EXACT = np.sqrt(2)

def objective_function(x):
    return np.sin(x) + np.cos(x)

def decode(bitstring, bounds, n_bits):
    integer_val = int("".join(map(str, bitstring)), 2)
    max_int = 2**n_bits - 1
    span = bounds[1] - bounds[0]
    return bounds[0] + (integer_val / max_int) * span

def crossover(p1, p2, rate):
    c1, c2 = p1.copy(), p2.copy()
    if np.random.rand() < rate:
        pt = np.random.randint(1, len(p1) - 1)
        c1 = np.concatenate((p1[:pt], p2[pt:]))
        c2 = np.concatenate((p2[:pt], p1[pt:]))
    return c1, c2

def mutation(bitstring, rate):
    for i in range(len(bitstring)):
        if np.random.rand() < rate:
            bitstring[i] = 1 - bitstring[i]
    return bitstring

# --- CC PHNG PHP LA CHN ---

def selection_roulette_wheel(pop, fitnesses):
   # La chn Bnh xe Roulette ( dng  cc bi trc).
    min_fitness = np.min(fitnesses)
    shifted_fitnesses = fitnesses - min_fitness
    total_fitness = np.sum(shifted_fitnesses)
    if total_fitness == 0:
        indices = np.random.randint(0, len(pop), size=len(pop))
    else:
        probabilities = shifted_fitnesses / total_fitness
        indices = np.random.choice(len(pop), size=len(pop), p=probabilities)
    return pop[indices]

def selection_tournament(pop, fitnesses, k=3):
    # La chn gii u (Tournament Selection).
    # - k: Kch thc ca mi "gii u".
    selected_pop = []
    pop_size = len(pop)
    for _ in range(pop_size):
        # 1. Chn ngu nhin k c th tham gia gii u
        tournament_indices = np.random.randint(0, pop_size, k)
        
        # 2. Tm fitness ca cc c th ny
        tournament_fitnesses = fitnesses[tournament_indices]
        
        # 3. Tm ra ngi chin thng (c fitness cao nht)
        winner_local_idx = np.argmax(tournament_fitnesses)
        winner_global_idx = tournament_indices[winner_local_idx]
        
        # 4. Thm ngi chin thng vo nhm c chn
        selected_pop.append(pop[winner_global_idx])
        
    return np.array(selected_pop)

def selection_random(pop, fitnesses):
    # La chn hon ton ngu nhin, khng quan tm n fitness.
    indices = np.random.randint(0, len(pop), size=len(pop))
    return pop[indices]

# --- HM CHY TH NGHIM  C CI TIN ---
def run_ga_experiment(pop_size, generations, crossover_rate, mutation_rate, selection_func):
    # Chy GA vi mt hm la chn c th c truyn vo.
    population = np.random.randint(0, 2, size=(pop_size, NUM_BITS))
    best_fitness_history = []
    best_h = -np.inf

    for _ in range(generations):
        decoded_pop = np.array([decode(p, BOUNDS, NUM_BITS) for p in population])
        fitnesses = objective_function(decoded_pop)
        
        current_best_h = np.max(fitnesses)
        if current_best_h > best_h:
            best_h = current_best_h
        best_fitness_history.append(best_h)
        
        # S DNG HM LA CHN C TRUYN VO
        mating_pool = selection_func(population, fitnesses)
        
        new_population = []
        for i in range(0, pop_size, 2):
            p1, p2 = mating_pool[i], mating_pool[i+1]
            c1, c2 = crossover(p1, p2, crossover_rate)
            c1 = mutation(c1, mutation_rate)
            c2 = mutation(c2, mutation_rate)
            new_population.extend([c1, c2])
            
        population = np.array(new_population)
        
    return best_fitness_history

# --- HM CHNH  SO SNH CC PHNG PHP ---
def compare_selection_methods(n_runs=10):
    BASE_PARAMS = {
        'pop_size': 50,
        'generations': 50,
        'crossover_rate': 0.8,
        'mutation_rate': 0.01,
    }

    # nh ngha cc phng php cn so snh
    # Dng partial  c nh tham s k cho hm tournament
    tournament_k3_func = partial(selection_tournament, k=3)
    
    methods_to_compare = {
        "Roulette Wheel": selection_roulette_wheel,
        "Tournament (k=3)": tournament_k3_func,
        "Random": selection_random
    }
    
    plt.figure(figsize=(12, 8))
    
    for name, func in methods_to_compare.items():
        print(f"--- ang th nghim phng php: {name} ---")
        all_histories = []
        for i in range(n_runs):
            #print(f"  Run {i+1}/{n_runs}...")
            history = run_ga_experiment(**BASE_PARAMS, selection_func=func)
            all_histories.append(history)
        
        avg_history = np.mean(np.array(all_histories), axis=0)
        plt.plot(avg_history, label=name, lw=2)

    plt.axhline(y=H_EXACT, color='black', linestyle='--', label=f'Gi tr Max chnh xc  {H_EXACT:.4f}')
    plt.title('So snh hiu qu cc phng php la chn', fontsize=16)
    plt.xlabel("Th h")
    plt.ylabel("Gi tr h(x) trung bnh tt nht")
    plt.legend()
    plt.grid(True)
    plt.ylim(0, 1.5) # Gii hn trc y  d so snh
    plt.show()

# Chy so snh
if __name__ == "__main__":
    compare_selection_methods(n_runs=10)
````

# Bi tp 5
# Trc quan ha qun th  V biu  phn tn (scatter plot) th hin s phn b ca qun th qua cc th h

---

##  Mc tiu

- Trc quan ha s phn b ca qun th c th trong qu trnh tin ha ca thut ton di truyn.
- Dng biu  phn tn (scatter plot)  th hin v tr cc c th trn khng gian bin s qua tng th h.
- Gip quan st r tin trnh hi t, a dng di truyn, v s thay i qun th theo thi gian.

---

##  tng thc hin

- Vi bi ton ti u hm $k(x, y, z) = x^2 + y^2 + z^2$, khng gian tm kim l 3 chiu.
- Mi c th l mt im trong khng gian $(x, y, z)$.
-  mi th h, ly v tr gii m ca ton b c th, v scatter plot.
- Nu khng th trc quan 3D trc tip, c th v scatter 2D tng cp bin: $(x, y)$, $(y, z)$, $(x, z)$ qua cc th h.

---

````python
#code vo y
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

# --- 1. CC THAM S CA THUT TON DI TRUYN ---

# Tham s ca bi ton
N_VARS = 3  # S lng bin (x, y, z)
BOUNDS = [-5, 5]  # Bin gi tr cho mi bin

# Tham s ca thut ton
POP_SIZE = 100         # Kch thc qun th
N_GENERATIONS = 50     # S th h
BITS_PER_VAR = 16      # S bit cho mi bin (quyt nh  chnh xc)
GENOME_LENGTH = N_VARS * BITS_PER_VAR
P_CROSSOVER = 0.8      # T l lai ghp
P_MUTATION = 0.01      # T l t bin

# --- 2. CC HM CT LI CA GA ---

# Hm gii m: Chuyn chui bit thnh cc gi tr (x, y, z)
def decode(genome, bounds, bits_per_var, n_vars):
    decoded_values = []
    min_bound, max_bound = bounds
    
    for i in range(n_vars):
        # Tch chui bit cho tng bin
        start = i * bits_per_var
        end = start + bits_per_var
        sub_genome = genome[start:end]
        
        # Chuyn t nh phn sang s nguyn
        int_val = int("".join(sub_genome.astype(str)), 2)
        
        # Chuyn v khong gi tr [min_bound, max_bound]
        max_int_val = 2**bits_per_var - 1
        value = min_bound + (int_val / max_int_val) * (max_bound - min_bound)
        decoded_values.append(value)
        
    return decoded_values

# Hm thch nghi (Fitness): nh gi mc  tt ca mt c th
def fitness_function(individual_coords):
    x, y, z = individual_coords
    k = x**2 + y**2 + z**2
    # Chuyn bi ton min -> max. Thm 1  trnh chia cho 0.
    return 1 / (1 + k)

# Hm chn lc (Selection): Dng phng php bnh xe Roulette
def selection(population, fitnesses):
    total_fitness = np.sum(fitnesses)
    probabilities = fitnesses / total_fitness
    # Chn 2 c th cha m
    selected_indices = np.random.choice(len(population), size=2, p=probabilities)
    return population[selected_indices[0]], population[selected_indices[1]]

# Hm lai ghp (Crossover): Lai ghp 1 im ct
def crossover(parent1, parent2, p_crossover):
    if np.random.rand() < p_crossover:
        # Chn ngu nhin im ct
        crossover_point = np.random.randint(1, len(parent1) - 1)
        child1 = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
        child2 = np.concatenate((parent2[:crossover_point], parent1[crossover_point:]))
        return child1, child2
    return parent1, parent2

# Hm t bin (Mutation): o bit
def mutation(individual, p_mutation):
    for i in range(len(individual)):
        if np.random.rand() < p_mutation:
            individual[i] = 1 - individual[i] # o bit (0 -> 1, 1 -> 0)
    return individual

# --- 3. VNG LP CHNH CA THUT TON ---

# Khi to qun th ngu nhin
population = np.random.randint(0, 2, size=(POP_SIZE, GENOME_LENGTH))

# Lu li lch s ta  ca qun th qua cc th h  v
population_history = []

print("Bt u qu trnh tin ha...")
for gen in range(N_GENERATIONS):
    # Gii m v tnh ton  thch nghi
    decoded_population = np.array([decode(ind, BOUNDS, BITS_PER_VAR, N_VARS) for ind in population])
    fitnesses = np.array([fitness_function(coords) for coords in decoded_population])
    
    # Lu li ta  ca th h hin ti
    population_history.append(decoded_population)
    
    # In thng tin th h hin ti
    best_idx = np.argmax(fitnesses)
    best_fitness = fitnesses[best_idx]
    best_solution = decoded_population[best_idx]
    k_value = best_solution[0]**2 + best_solution[1]**2 + best_solution[2]**2
    
    if (gen % 5 == 0 or gen == N_GENERATIONS - 1):
        print(f"Th h {gen}: Best Fitness = {best_fitness:.5f}, "
              f"Best Solution (x,y,z) = ({best_solution[0]:.4f}, {best_solution[1]:.4f}, {best_solution[2]:.4f}), "
              f"k(x,y,z) = {k_value:.6f}")

    # To th h mi
    new_population = []
    for _ in range(POP_SIZE // 2):
        parent1, parent2 = selection(population, fitnesses)
        child1, child2 = crossover(parent1, parent2, P_CROSSOVER)
        child1 = mutation(child1, P_MUTATION)
        child2 = mutation(child2, P_MUTATION)
        new_population.extend([child1, child2])
        
    population = np.array(new_population)

print("Qu trnh tin ha hon tt.")


# --- 4. TRC QUAN HA BNG BIU  NG ---

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# im ti u (0,0,0)  tham chiu
ax.scatter([0], [0], [0], c='red', marker='*', s=200, label='im ti u (0,0,0)')

# Hm cp nht cho mi frame ca animation
def update(frame):
    # Xa cc im ca frame trc
    ax.cla()
    
    # Ly d liu ca th h hin ti (frame)
    current_pop_coords = population_history[frame]
    x_coords = current_pop_coords[:, 0]
    y_coords = current_pop_coords[:, 1]
    z_coords = current_pop_coords[:, 2]
    
    # V cc c th ca th h hin ti
    ax.scatter(x_coords, y_coords, z_coords, c='blue', alpha=0.6)
    
    # Gi li im ti u v ci t cho trc
    ax.scatter([0], [0], [0], c='red', marker='*', s=200, label='im ti u (0,0,0)')
    ax.set_title(f"Phn b qun th  th h {frame}")
    ax.set_xlabel('Bin x')
    ax.set_ylabel('Bin y')
    ax.set_zlabel('Bin z')
    ax.set_xlim(BOUNDS)
    ax.set_ylim(BOUNDS)
    ax.set_zlim(BOUNDS)
    ax.legend()
    ax.grid(True)

# To v hin th animation
ani = FuncAnimation(fig, update, frames=N_GENERATIONS, interval=200, repeat=False)
plt.show()
````
