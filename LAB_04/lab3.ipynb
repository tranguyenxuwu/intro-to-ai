## Lý thuyết
Thuật toán di truyền (Genetic Algorithm - GA) là một phương pháp tìm kiếm dựa trên cơ chế tiến hóa tự nhiên, lấy cảm hứng từ thuyết tiến hóa của Darwin. GA được sử dụng để tìm lời giải gần tối ưu cho các bài toán tối ưu hóa phức tạp.

### Các bước chính của thuật toán:
1. **Khởi tạo quần thể**: Tạo ngẫu nhiên một tập hợp các cá thể (giải pháp tiềm năng).
2. **Đánh giá độ thích nghi (Fitness Evaluation)**: Tính giá trị hàm mục tiêu cho mỗi cá thể.
3. **Lựa chọn (Selection)**: Chọn các cá thể tốt nhất để lai ghép (thường sử dụng các phương pháp như Tournament Selection hoặc Roulette Wheel).
4. **Lai ghép (Crossover)**: Kết hợp hai cá thể để tạo ra cá thể mới.
5. **Đột biến (Mutation)**: Thay đổi ngẫu nhiên một số gen để duy trì tính đa dạng.
6. **Lặp lại**: Quay lại bước 2 cho đến khi đạt điều kiện dừng (số thế hệ hoặc giá trị fitness mong muốn).

## Ví dụ 1: Tối ưu hóa hàm một biến

### Bài toán  
Tìm $x$ sao cho  
$$
f(x) = -(x^2) + 10x + 50
$$  
đạt giá trị lớn nhất trong khoảng  
$$
x \in [0, 10].
$$

# Mã giả
```plaintext
Tạo quần thể mới rỗng
Lặp cho đến khi quần thể mới đủ N cá thể:
    Chọn cha mẹ 1 bằng Tournament Selection:
        - Chọn ngẫu nhiên tournament_size cá thể trong quần thể
        - Chọn cá thể có fitness cao nhất làm cha mẹ 1
    Chọn cha mẹ 2 bằng Tournament Selection tương tự
    
    Thực hiện lai ghép (crossover) giữa cha mẹ 1 và cha mẹ 2 với xác suất crossover_rate:
        Nếu crossover thành công:
            Tạo con mới có giá trị trung bình của cha mẹ
        Ngược lại:
            Con mới là cha mẹ 1
            
    Thực hiện đột biến cho con mới với xác suất mutation_rate:
        Nếu đột biến:
            Gán con mới giá trị ngẫu nhiên trong [min_val, max_val]
            
    Thêm con mới vào quần thể mới
    
Cập nhật quần thể hiện tại = quần thể mới
In ra cá thể tốt nhất và giá trị fitness tương ứng của thế hệ


````python
import numpy as np
import random
import matplotlib.pyplot as plt

# Hàm mục tiêu
def fitness_function(x):
    return -(x**2) + 10*x + 50

# Khởi tạo quần thể
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# Lựa chọn (Tournament Selection)
def select_parents(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghép
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# Đột biến
def mutate(individual, mutation_rate=0.1, min_val=0, max_val=10):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thuật toán di truyền
def genetic_algorithm_example1(pop_size=50, generations=100, min_val=0, max_val=10):
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents(population, fitness_values)
            parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function(x) for x in population])]
        print(f'Thế hệ {generation + 1}: x = {best_individual:.4f}, f(x) = {best_fitness:.4f}')
    
    best_fitness = max([fitness_function(x) for x in population])
    best_individual = population[np.argmax([fitness_function(x) for x in population])]
    
    # Trực quan hóa
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị f(x) tốt nhất')
    plt.title('Sự thay đổi của f(x) qua các thế hệ')
    plt.show()
    
    return best_individual, best_fitness

# Chạy thuật toán
best_x, best_f = genetic_algorithm_example1()
print(f'\nKết quả cuối cùng: x = {best_x:.4f}, f(x) = {best_f:.4f}')
````

### Mã giả 

```plaintext
Tạo quần thể mới rỗng
Lặp cho đến khi quần thể mới đủ N cá thể:
    Chọn cha mẹ 1 bằng Tournament Selection trên quần thể hiện tại
    Chọn cha mẹ 2 bằng Tournament Selection trên quần thể hiện tại
    Thực hiện lai ghép (crossover) giữa cha mẹ 1 và cha mẹ 2 với xác suất crossover_rate:
        Nếu crossover thành công:
            Tạo con mới bằng trung bình giá trị (x, y) của cha mẹ
        Ngược lại:
            Con mới là cha mẹ 1
    Thực hiện đột biến cho con mới với xác suất mutation_rate từng biến:
        Nếu đột biến:
            Gán giá trị ngẫu nhiên trong [min_val, max_val] cho biến tương ứng
    Thêm con mới vào quần thể mới

Cập nhật quần thể hiện tại = quần thể mới
In kết quả tốt nhất thế hệ

````python
# Hàm mục tiêu
def fitness_function_2d(x, y):
    return -(x**2 + y**2)  # Sử dụng dấu âm vì GA tìm max

# Khởi tạo quần thể
def initialize_population_2d(pop_size, min_val, max_val):
    return [(random.uniform(min_val, max_val), random.uniform(min_val, max_val)) for _ in range(pop_size)]

# Lựa chọn
def select_parents_2d(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghép
def crossover_2d(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return ((parent1[0] + parent2[0]) / 2, (parent1[1] + parent2[1]) / 2)
    return parent1

# Đột biến
def mutate_2d(individual, mutation_rate=0.1, min_val=-5, max_val=5):
    x, y = individual
    if random.random() < mutation_rate:
        x = random.uniform(min_val, max_val)
    if random.random() < mutation_rate:
        y = random.uniform(min_val, max_val)
    return (x, y)

# Thuật toán di truyền
def genetic_algorithm_example2(pop_size=50, generations=100, min_val=-5, max_val=5):
    population = initialize_population_2d(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function_2d(x, y) for x, y in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(-best_fitness)  # Chuyển về giá trị thực
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents_2d(population, fitness_values)
            parent2 = select_parents_2d(population, fitness_values)
            child = crossover_2d(parent1, parent2)
            child = mutate_2d(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
        print(f'Thế hệ {generation + 1}: (x, y) = ({best_individual[0]:.4f}, {best_individual[1]:.4f}), g(x, y) = {-best_fitness:.4f}')
    
    best_fitness = max([fitness_function_2d(x, y) for x, y in population])
    best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
    
    # Trực quan hóa
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị g(x, y) tốt nhất')
    plt.title('Sự thay đổi của g(x, y) qua các thế hệ')
    plt.show()
    
    return best_individual, -best_fitness

# Chạy thuật toán
best_xy, best_g = genetic_algorithm_example2()
print(f'\nKết quả cuối cùng: (x, y) = ({best_xy[0]:.4f}, {best_xy[1]:.4f}), g(x, y) = {best_g:.4f}')
````

# Bài tập 1
# Thay đổi hàm mục tiêu**: Thử áp dụng thuật toán cho hàm $h(x) = sin(x) + cos(x)$ trong khoảng $x \in [0, 2\pi]$. So sánh kết quả với nghiệm chính xác.

## Tối ưu hàm $h(x) = \sin(x) + \cos(x)$ bằng Thuật Toán Di Truyền (Genetic Algorithm)

## Mục tiêu
- Tối ưu hàm:
  $$
  h(x) = \sin(x) + \cos(x)
  $$
  trong miền:
  $$
  x \in [0, 2\pi]
  $$
- Tìm giá trị **x** sao cho **h(x)** đạt giá trị lớn nhất.

---

## Phân tích nghiệm chính xác

Ta có:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xảy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

## Ý tưởng thuật toán di truyền (GA)

1. **Mã hóa**: Biểu diễn giá trị \( x \) bằng chuỗi nhị phân dài 16 bit.  
2. **Giải mã**: Chuyển từ chuỗi nhị phân sang giá trị thực trong khoảng \([0, 2\pi]\).  
3. **Fitness**: Sử dụng chính giá trị \( h(x) \) làm độ thích nghi.  
4. **Chọn lọc**: Roulette Wheel Selection.  
5. **Lai ghép**: Lai 1 điểm với xác suất cao.  
6. **Đột biến**: Lật bit với xác suất thấp.

---


````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# code vào  đây nhe


# --- 1. Cài đặt các tham số cho thuật toán ---
# Miền giá trị của x
BOUNDS = [0, 2 * np.pi]
# Số lượng bit để mã hóa một cá thể (giá trị x)
NUM_BITS = 16
# Kích thước quần thể
POP_SIZE = 50
# Tỷ lệ lai ghép
CROSSOVER_RATE = 0.8
# Tỷ lệ đột biến
MUTATION_RATE = 0.01
# Số thế hệ
NUM_GENERATIONS = 100

# --- 2. Định nghĩa các hàm cốt lõi ---

def objective_function(x):
    #Hàm mục tiêu h(x) cần tối ưu.#
    return np.sin(x) + np.cos(x)

def decode(bitstring, bounds, n_bits):
    #Giải mã chuỗi bit thành giá trị thực x trong miền [min, max].#
    # Chuyển chuỗi bit thành số nguyên
    integer_val = int("".join(map(str, bitstring)), 2)
    
    # Tính giá trị lớn nhất mà số nguyên có thể đạt được
    max_int = 2**n_bits - 1
    
    # Ánh xạ giá trị nguyên sang miền giá trị thực
    span = bounds[1] - bounds[0]
    decoded_val = bounds[0] + (integer_val / max_int) * span
    return decoded_val

def selection(pop, fitnesses): # Chọn lọc cá thể bằng phương pháp Roulette Wheel.
    # Dịch chuyển fitness để tất cả đều không âm
    min_fitness = np.min(fitnesses)
    shifted_fitnesses = fitnesses - min_fitness
    
    total_fitness = np.sum(shifted_fitnesses)
    
    # Nếu tất cả các fitness bằng nhau, trả về lựa chọn ngẫu nhiên
    if total_fitness == 0:
        return pop[np.random.randint(0, len(pop), size=len(pop))]

    # Tính xác suất chọn lọc
    probabilities = shifted_fitnesses / total_fitness
    
    # Chọn lọc các cá thể mới
    indices = np.random.choice(len(pop), size=len(pop), p=probabilities)
    return pop[indices]

# Lai ghép 1 điểm
def crossover(p1, p2, rate):
    c1, c2 = p1.copy(), p2.copy()
    if np.random.rand() < rate:
        # Chọn ngẫu nhiên 1 điểm để lai ghép
        pt = np.random.randint(1, len(p1) - 1)
        # Thực hiện lai ghép
        c1 = np.concatenate((p1[:pt], p2[pt:]))
        c2 = np.concatenate((p2[:pt], p1[pt:]))
    return c1, c2

def mutation(bitstring, rate):
    #Thực hiện đột biến bằng cách lật bit.#
    for i in range(len(bitstring)):
        if np.random.rand() < rate:
            # Lật bit (0 -> 1, 1 -> 0)
            bitstring[i] = 1 - bitstring[i]
    return bitstring

# --- 3. Chạy thuật toán di truyền ---

# Khởi tạo quần thể ban đầu
population = np.random.randint(0, 2, size=(POP_SIZE, NUM_BITS))

# Lưu lại cá thể tốt nhất qua các thế hệ
best_x, best_h = 0, -np.inf
best_fitness_history = []

print("Bắt đầu quá trình tối ưu...")
for gen in range(NUM_GENERATIONS):
    # 1. Giải mã và đánh giá (tính fitness)
    decoded_pop = np.array([decode(p, BOUNDS, NUM_BITS) for p in population])
    fitnesses = objective_function(decoded_pop)

    # 2. Tìm cá thể tốt nhất trong thế hệ hiện tại và toàn cục
    current_best_idx = np.argmax(fitnesses)
    if fitnesses[current_best_idx] > best_h:
        best_h = fitnesses[current_best_idx]
        best_x = decoded_pop[current_best_idx]
        
    best_fitness_history.append(best_h)
    
    if (gen + 1) % 10 == 0:
        print(f"> Thế hệ {gen+1:3d}, x tốt nhất = {best_x:.4f}, h(x) = {best_h:.4f}")

    # 3. Chọn lọc
    mating_pool = selection(population, fitnesses)
    
    # 4. Tạo thế hệ mới
    new_population = []
    for i in range(0, POP_SIZE, 2):
        p1, p2 = mating_pool[i], mating_pool[i+1]
        
        # 5. Lai ghép
        c1, c2 = crossover(p1, p2, CROSSOVER_RATE)
        
        # 6. Đột biến
        c1 = mutation(c1, MUTATION_RATE)
        c2 = mutation(c2, MUTATION_RATE)
        
        new_population.append(c1)
        new_population.append(c2)
        
    population = np.array(new_population)

# --- 4. In kết quả và so sánh ---
print("\n--- KẾT QUẢ CUỐI CÙNG ---")

# Nghiệm chính xác
x_exact = np.pi / 4
h_exact = np.sqrt(2)

print(f"Nghiệm tìm thấy bởi GA: ")
print(f"  x = {best_x:.6f}")
print(f"  h(x) = {best_h:.6f}")
print("-" * 20)
print(f"Nghiệm chính xác (lý thuyết):")
print(f"  x = {x_exact:.6f} (π/4)")
print(f"  h(x) = {h_exact:.6f} (√2)")
print("-" * 20)
print(f"Sai số tuyệt đối cho x: {abs(best_x - x_exact):.6f}")
print(f"Sai số tuyệt đối cho h(x): {abs(best_h - h_exact):.6f}")

# --- 5. Trực quan hóa ---
plt.figure(figsize=(10, 6))
plt.plot(best_fitness_history, label='Giá trị h(x) tốt nhất mỗi thế hệ')
plt.axhline(y=h_exact, color='r', linestyle='--', label=f'Giá trị Max chính xác (√2 ≈ {h_exact:.4f})')
plt.xlabel("Thế hệ")
plt.ylabel("Giá trị h(x) (Fitness)")
plt.title("Sự hội tụ của Thuật toán Di truyền")
plt.legend()
plt.grid(True)
plt.show()
````

##  Ý tưởng thuật toán di truyền (GA)

| Thành phần        | Chi tiết                                                                 |
|------------------|-------------------------------------------------------------------------|
| **Mã hóa**        | Dùng chuỗi nhị phân 16 bit để biểu diễn giá trị $x \in [0, 2\pi]$         |
| **Giải mã**       | Ánh xạ nhị phân sang giá trị thực: $x = 0 + \dfrac{\text{decimal}}{2^{16}-1} \cdot 2\pi$ |
| **Fitness**       | Chính là $h(x) = \sin(x) + \cos(x)$                                     |
| **Chọn lọc**      | Roulette Wheel Selection                                                |
| **Lai ghép**      | 1 điểm cắt với xác suất 0.8                                              |
| **Đột biến**      | Xác suất 5% (0.05)                                                       |

---

````python


# --- 1. Cài đặt các tham số cho thuật toán ---
# Miền giá trị của x
BOUNDS = [0, 2 * np.pi]
# Số lượng bit để mã hóa một cá thể (giá trị x)
NUM_BITS = 16
# Kích thước quần thể
POP_SIZE = 50
# Tỷ lệ lai ghép
CROSSOVER_RATE = 0.8
# Tỷ lệ đột biến (THEO YÊU CẦU BÀI TẬP)
MUTATION_RATE = 0.05
# Số thế hệ
NUM_GENERATIONS = 100

# --- 2. Định nghĩa các hàm cốt lõi ---

def objective_function(x):
    #Hàm mục tiêu h(x) cần tối ưu.#
    return np.sin(x) + np.cos(x)

def decode(bitstring, bounds, n_bits):
    #Giải mã chuỗi bit thành giá trị thực x.#
    # Chuyển chuỗi bit thành số nguyên
    integer_val = int("".join(map(str, bitstring)), 2)
    # Giá trị lớn nhất mà số nguyên có thể đạt được
    max_int = 2**n_bits - 1
    # Ánh xạ giá trị nguyên sang miền giá trị thực
    span = bounds[1] - bounds[0]
    decoded_val = bounds[0] + (integer_val / max_int) * span
    return decoded_val

def selection(pop, fitnesses):
    #Chọn lọc cá thể bằng phương pháp Roulette Wheel.#
    # Dịch chuyển fitness để tất cả đều không âm
    min_fitness = np.min(fitnesses)
    shifted_fitnesses = fitnesses - min_fitness
    
    total_fitness = np.sum(shifted_fitnesses)
    
    # Nếu tất cả các fitness bằng nhau, trả về lựa chọn ngẫu nhiên
    if total_fitness == 0:
        return pop[np.random.randint(0, len(pop), size=len(pop))]

    probabilities = shifted_fitnesses / total_fitness
    indices = np.random.choice(len(pop), size=len(pop), p=probabilities)
    return pop[indices]

def crossover(p1, p2, rate):
    #Thực hiện lai ghép 1 điểm.#
    c1, c2 = p1.copy(), p2.copy()
    if np.random.rand() < rate:
        pt = np.random.randint(1, len(p1) - 1)
        c1 = np.concatenate((p1[:pt], p2[pt:]))
        c2 = np.concatenate((p2[:pt], p1[pt:]))
    return c1, c2

def mutation(bitstring, rate):
    #Thực hiện đột biến bằng cách lật bit.#
    for i in range(len(bitstring)):
        if np.random.rand() < rate:
            bitstring[i] = 1 - bitstring[i]
    return bitstring

# --- 3. Chạy thuật toán di truyền ---

# Khởi tạo quần thể ban đầu
population = np.random.randint(0, 2, size=(POP_SIZE, NUM_BITS))

# Lưu lại cá thể tốt nhất qua các thế hệ
best_x, best_h = 0, -np.inf
best_fitness_history = []

print("Bắt đầu quá trình tối ưu với Mutation Rate = 0.05...")
for gen in range(NUM_GENERATIONS):
    # 1. Giải mã và đánh giá
    decoded_pop = np.array([decode(p, BOUNDS, NUM_BITS) for p in population])
    fitnesses = objective_function(decoded_pop)

    # 2. Tìm cá thể tốt nhất
    current_best_idx = np.argmax(fitnesses)
    if fitnesses[current_best_idx] > best_h:
        best_h = fitnesses[current_best_idx]
        best_x = decoded_pop[current_best_idx]
        
    best_fitness_history.append(best_h)
    
    if (gen + 1) % 10 == 0:
        print(f"> Thế hệ {gen+1:3d}, x tốt nhất = {best_x:.4f}, h(x) = {best_h:.4f}")

    # 3. Chọn lọc
    mating_pool = selection(population, fitnesses)
    
    # 4. Tạo thế hệ mới (Lai ghép và Đột biến)
    new_population = []
    for i in range(0, POP_SIZE, 2):
        p1, p2 = mating_pool[i], mating_pool[i+1]
        c1, c2 = crossover(p1, p2, CROSSOVER_RATE)
        c1 = mutation(c1, MUTATION_RATE)
        c2 = mutation(c2, MUTATION_RATE)
        new_population.append(c1)
        new_population.append(c2)
        
    population = np.array(new_population)

# --- 4. In kết quả và so sánh ---
print("\n--- KẾT QUẢ CUỐI CÙNG ---")

# Nghiệm chính xác
x_exact = np.pi / 4
h_exact = np.sqrt(2)

print(f"Nghiệm tìm thấy bởi GA: ")
print(f"  x = {best_x:.6f}")
print(f"  h(x) = {best_h:.6f}")
print("-" * 20)
print(f"Nghiệm chính xác (lý thuyết):")
print(f"  x = {x_exact:.6f} (π/4)")
print(f"  h(x) = {h_exact:.6f} (√2)")
print("-" * 20)
print(f"Sai số tuyệt đối cho x: {abs(best_x - x_exact):.6f}")
print(f"Sai số tuyệt đối cho h(x): {abs(best_h - h_exact):.6f}")

# --- 5. Trực quan hóa ---
plt.figure(figsize=(10, 6))
plt.plot(best_fitness_history, label=f'Giá trị h(x) tốt nhất (Mutation Rate = {MUTATION_RATE})')
plt.axhline(y=h_exact, color='r', linestyle='--', label=f'Giá trị Max chính xác (√2 ≈ {h_exact:.4f})')
plt.xlabel("Thế hệ")
plt.ylabel("Giá trị h(x) (Fitness)")
plt.title("Sự hội tụ của Thuật toán Di truyền")
plt.legend()
plt.grid(True)
plt.show()
````

# Thay đổi tham số thuật toán di truyền và phân tích tốc độ hội tụ

**Mục tiêu:** Thử thay đổi các tham số chính của thuật toán di truyền gồm:

- Kích thước quần thể: `pop_size`
- Số thế hệ: `generations`
- Tỷ lệ đột biến: `mutation_rate`
- Tỷ lệ lai ghép: `crossover_rate`

và phân tích ảnh hưởng của từng tham số đến tốc độ hội tụ của thuật toán.

---

## Các tham số và vai trò

| Tham số         | Ý nghĩa                                  | Ảnh hưởng chính                                                                                         |
|-----------------|-----------------------------------------|--------------------------------------------------------------------------------------------------------|
| `pop_size`      | Số cá thể trong quần thể                 | Quần thể lớn giúp khám phá không gian tìm kiếm tốt hơn nhưng tốn nhiều thời gian tính toán. Quần thể nhỏ hội tụ nhanh nhưng dễ bị kẹt cực trị cục bộ. |
| `generations`   | Số vòng lặp của thuật toán               | Số thế hệ càng nhiều thì thuật toán có nhiều cơ hội cải thiện nghiệm, tuy nhiên thời gian chạy cũng tăng. |
| `mutation_rate` | Xác suất đột biến mỗi bit                | Đột biến cao giúp tăng đa dạng gene, tránh bị kẹt trong cực trị cục bộ, nhưng nếu quá cao làm quá trình tìm kiếm nhiễu loạn. |
| `crossover_rate`| Xác suất lai ghép giữa hai cá thể       | Lai ghép giúp phối hợp gen tốt, tăng khả năng tìm nghiệm tốt. Nếu quá thấp, quần thể kém đa dạng; quá cao có thể phá hủy cấu trúc tốt của cá thể. |

---

## Phân tích ảnh hưởng

- **`pop_size`**: Kích thước quần thể lớn giúp tìm kiếm toàn cục tốt hơn nhưng chậm hơn, kích thước nhỏ làm thuật toán hội tụ nhanh nhưng dễ mắc kẹt.
- **`generations`**: Số thế hệ lớn tăng khả năng cải tiến kết quả nhưng làm tăng thời gian tính toán.
- **`mutation_rate`**: Tỷ lệ đột biến vừa phải giúp duy trì đa dạng di truyền, tránh kẹt cục bộ; quá cao hoặc quá thấp đều làm giảm hiệu quả.
- **`crossover_rate`**: Tỷ lệ lai ghép hợp lý giúp khai thác tốt thông tin di truyền, tăng hiệu quả hội tụ.

---




````python
# code vào đây 

# --- HÀM CỐT LÕI CỦA GA (Giữ nguyên từ các bài trước) ---
BOUNDS = [0, 2 * np.pi]
NUM_BITS = 16
H_EXACT = np.sqrt(2)

def objective_function(x):
    return np.sin(x) + np.cos(x)

def decode(bitstring, bounds, n_bits):
    integer_val = int("".join(map(str, bitstring)), 2)
    max_int = 2**n_bits - 1
    span = bounds[1] - bounds[0]
    return bounds[0] + (integer_val / max_int) * span

def selection(pop, fitnesses):
    min_fitness = np.min(fitnesses)
    shifted_fitnesses = fitnesses - min_fitness
    total_fitness = np.sum(shifted_fitnesses)
    if total_fitness == 0:
        return pop[np.random.randint(0, len(pop), size=len(pop))]
    probabilities = shifted_fitnesses / total_fitness
    indices = np.random.choice(len(pop), size=len(pop), p=probabilities)
    return pop[indices]

def crossover(p1, p2, rate):
    c1, c2 = p1.copy(), p2.copy()
    if np.random.rand() < rate:
        pt = np.random.randint(1, len(p1) - 1)
        c1 = np.concatenate((p1[:pt], p2[pt:]))
        c2 = np.concatenate((p2[:pt], p1[pt:]))
    return c1, c2

def mutation(bitstring, rate):
    for i in range(len(bitstring)):
        if np.random.rand() < rate:
            bitstring[i] = 1 - bitstring[i]
    return bitstring

# --- HÀM CHẠY THÍ NGHIỆM ---
def run_ga_experiment(pop_size, generations, crossover_rate, mutation_rate):
    
    # Chạy một lượt thuật toán GA với các tham số cho trước.
    # Trả về lịch sử giá trị fitness tốt nhất qua các thế hệ.
    population = np.random.randint(0, 2, size=(pop_size, NUM_BITS))
    best_fitness_history = []
    best_h = -np.inf

    for _ in range(generations):
        decoded_pop = np.array([decode(p, BOUNDS, NUM_BITS) for p in population])
        fitnesses = objective_function(decoded_pop)
        
        current_best_h = np.max(fitnesses)
        if current_best_h > best_h:
            best_h = current_best_h
        best_fitness_history.append(best_h)
        
        mating_pool = selection(population, fitnesses)
        
        new_population = []
        for i in range(0, pop_size, 2):
            p1, p2 = mating_pool[i], mating_pool[i+1]
            c1, c2 = crossover(p1, p2, crossover_rate)
            c1 = mutation(c1, mutation_rate)
            c2 = mutation(c2, mutation_rate)
            new_population.extend([c1, c2])
            
        population = np.array(new_population)
        
    return best_fitness_history

# --- HÀM ĐỂ CHẠY VÀ VẼ BIỂU ĐỒ ---
def analyze_parameter(param_name, param_values, base_params, n_runs=10):
    
    #Phân tích ảnh hưởng của một tham số bằng cách chạy nhiều lần và vẽ biểu đồ.
    plt.figure(figsize=(12, 7))
    plt.title(f'Phân tích ảnh hưởng của {param_name}', fontsize=16)
    
    for value in param_values:
        print(f"--- Đang thí nghiệm: {param_name} = {value} ---")
        current_params = base_params.copy()
        current_params[param_name] = value
        
        all_histories = []
        for i in range(n_runs):
           # print(f"  Run {i+1}/{n_runs}...")
            history = run_ga_experiment(**current_params)
            all_histories.append(history)
            
        # Lấy trung bình kết quả của n_runs lần chạy
        avg_history = np.mean(np.array(all_histories), axis=0)
        
        plt.plot(avg_history, label=f'{param_name} = {value}')

    plt.axhline(y=H_EXACT, color='r', linestyle='--', label=f'Giá trị Max chính xác ≈ {H_EXACT:.4f}')
    plt.xlabel("Thế hệ")
    plt.ylabel("Giá trị h(x) trung bình tốt nhất")
    plt.legend()
    plt.grid(True)
    plt.show()

# --- THỰC HIỆN CÁC THÍ NGHIỆM ---
if __name__ == "__main__":
    BASE_PARAMETERS = {
        'pop_size': 50,
        'generations': 50,
        'crossover_rate': 0.8,
        'mutation_rate': 0.01
    }

    # 1. Phân tích Kích thước quần thể (pop_size)
    analyze_parameter(
        param_name='pop_size',
        param_values=[10, 50, 200],
        base_params=BASE_PARAMETERS
    )
    
    # 2. Phân tích Tỷ lệ đột biến (mutation_rate)
    analyze_parameter(
        param_name='mutation_rate',
        param_values=[0.001, 0.01, 0.05],
        base_params=BASE_PARAMETERS
    )

    # 3. Phân tích Tỷ lệ lai ghép (crossover_rate)
    analyze_parameter(
        param_name='crossover_rate',
        param_values=[0.2, 0.8, 1.0],
        base_params=BASE_PARAMETERS
    )
````

|----------------------------|-----------------------------------------------------------------------------------------|------------------------------------|-----------------------------------|

| Tournament Selection        | Chọn ngẫu nhiên một số cá thể, chọn cá thể tốt nhất trong nhóm                          | Kiểm soát đa dạng tốt              | Cần chọn kích thước nhóm phù hợp  |

3. So sánh kết quả về:
   - Tốc độ hội tụ (số thế hệ cần để đạt nghiệm tốt).
   - Giá trị nghiệm tốt nhất đạt được.
4. Phân tích ưu – nhược điểm từng phương pháp trong thực tế.

---

##  Kết quả và phân tích

- Phương pháp **Roulette Wheel** thường giúp tăng tốc độ hội tụ nhờ ưu tiên cá thể tốt.
- Tuy nhiên có thể làm giảm đa dạng gen nếu cá thể ưu tú chiếm ưu thế quá lớn, dễ rơi vào cực trị cục bộ.
- Phương pháp **Tournament** kiểm soát tốt sự đa dạng nhưng có thể chậm hơn trong một số trường hợp.
- Phương pháp **Random** đơn giản nhưng ít hiệu quả trong tối ưu nâng cao.

---


````python
# code vào đây nhé
import numpy as np
import matplotlib.pyplot as plt
from functools import partial

# --- HÀM CỐT LÕI CỦA GA (Giữ nguyên) ---
BOUNDS = [0, 2 * np.pi]
NUM_BITS = 16
H_EXACT = np.sqrt(2)

def objective_function(x):
    return np.sin(x) + np.cos(x)

def decode(bitstring, bounds, n_bits):
    integer_val = int("".join(map(str, bitstring)), 2)
    max_int = 2**n_bits - 1
    span = bounds[1] - bounds[0]
    return bounds[0] + (integer_val / max_int) * span

def crossover(p1, p2, rate):
    c1, c2 = p1.copy(), p2.copy()
    if np.random.rand() < rate:
        pt = np.random.randint(1, len(p1) - 1)
        c1 = np.concatenate((p1[:pt], p2[pt:]))
        c2 = np.concatenate((p2[:pt], p1[pt:]))
    return c1, c2

def mutation(bitstring, rate):
    for i in range(len(bitstring)):
        if np.random.rand() < rate:
            bitstring[i] = 1 - bitstring[i]
    return bitstring

# --- CÁC PHƯƠNG PHÁP LỰA CHỌN ---

def selection_roulette_wheel(pop, fitnesses):
   # Lựa chọn Bánh xe Roulette (đã dùng ở các bài trước).
    min_fitness = np.min(fitnesses)
    shifted_fitnesses = fitnesses - min_fitness
    total_fitness = np.sum(shifted_fitnesses)
    if total_fitness == 0:
        indices = np.random.randint(0, len(pop), size=len(pop))
    else:
        probabilities = shifted_fitnesses / total_fitness
        indices = np.random.choice(len(pop), size=len(pop), p=probabilities)
    return pop[indices]

def selection_tournament(pop, fitnesses, k=3):
    # Lựa chọn giải đấu (Tournament Selection).
    # - k: Kích thước của mỗi "giải đấu".
    selected_pop = []
    pop_size = len(pop)
    for _ in range(pop_size):
        # 1. Chọn ngẫu nhiên k cá thể tham gia giải đấu
        tournament_indices = np.random.randint(0, pop_size, k)
        
        # 2. Tìm fitness của các cá thể này
        tournament_fitnesses = fitnesses[tournament_indices]
        
        # 3. Tìm ra người chiến thắng (có fitness cao nhất)
        winner_local_idx = np.argmax(tournament_fitnesses)
        winner_global_idx = tournament_indices[winner_local_idx]
        
        # 4. Thêm người chiến thắng vào nhóm được chọn
        selected_pop.append(pop[winner_global_idx])
        
    return np.array(selected_pop)

def selection_random(pop, fitnesses):
    # Lựa chọn hoàn toàn ngẫu nhiên, không quan tâm đến fitness.
    indices = np.random.randint(0, len(pop), size=len(pop))
    return pop[indices]

# --- HÀM CHẠY THÍ NGHIỆM ĐÃ ĐƯỢC CẢI TIẾN ---
def run_ga_experiment(pop_size, generations, crossover_rate, mutation_rate, selection_func):
    # Chạy GA với một hàm lựa chọn cụ thể được truyền vào.
    population = np.random.randint(0, 2, size=(pop_size, NUM_BITS))
    best_fitness_history = []
    best_h = -np.inf

    for _ in range(generations):
        decoded_pop = np.array([decode(p, BOUNDS, NUM_BITS) for p in population])
        fitnesses = objective_function(decoded_pop)
        
        current_best_h = np.max(fitnesses)
        if current_best_h > best_h:
            best_h = current_best_h
        best_fitness_history.append(best_h)
        
        # SỬ DỤNG HÀM LỰA CHỌN ĐƯỢC TRUYỀN VÀO
        mating_pool = selection_func(population, fitnesses)
        
        new_population = []
        for i in range(0, pop_size, 2):
            p1, p2 = mating_pool[i], mating_pool[i+1]
            c1, c2 = crossover(p1, p2, crossover_rate)
            c1 = mutation(c1, mutation_rate)
            c2 = mutation(c2, mutation_rate)
            new_population.extend([c1, c2])
            
        population = np.array(new_population)
        
    return best_fitness_history

# --- HÀM CHÍNH ĐỂ SO SÁNH CÁC PHƯƠNG PHÁP ---
def compare_selection_methods(n_runs=10):
    BASE_PARAMS = {
        'pop_size': 50,
        'generations': 50,
        'crossover_rate': 0.8,
        'mutation_rate': 0.01,
    }

    # Định nghĩa các phương pháp cần so sánh
    # Dùng partial để cố định tham số k cho hàm tournament
    tournament_k3_func = partial(selection_tournament, k=3)
    
    methods_to_compare = {
        "Roulette Wheel": selection_roulette_wheel,
        "Tournament (k=3)": tournament_k3_func,
        "Random": selection_random
    }
    
    plt.figure(figsize=(12, 8))
    
    for name, func in methods_to_compare.items():
        print(f"--- Đang thí nghiệm phương pháp: {name} ---")
        all_histories = []
        for i in range(n_runs):
            #print(f"  Run {i+1}/{n_runs}...")
            history = run_ga_experiment(**BASE_PARAMS, selection_func=func)
            all_histories.append(history)
        
        avg_history = np.mean(np.array(all_histories), axis=0)
        plt.plot(avg_history, label=name, lw=2)

    plt.axhline(y=H_EXACT, color='black', linestyle='--', label=f'Giá trị Max chính xác ≈ {H_EXACT:.4f}')
    plt.title('So sánh hiệu quả các phương pháp lựa chọn', fontsize=16)
    plt.xlabel("Thế hệ")
    plt.ylabel("Giá trị h(x) trung bình tốt nhất")
    plt.legend()
    plt.grid(True)
    plt.ylim(0, 1.5) # Giới hạn trục y để dễ so sánh
    plt.show()

# Chạy so sánh
if __name__ == "__main__":
    compare_selection_methods(n_runs=10)
````

# Trực quan hóa quần thể – Vẽ biểu đồ phân tán (scatter plot) thể hiện sự phân bố của quần thể qua các thế hệ

##  Mục tiêu

- Trực quan hóa sự phân bố của quần thể cá thể trong quá trình tiến hóa của thuật toán di truyền.
- Dùng biểu đồ phân tán (scatter plot) để thể hiện vị trí các cá thể trên không gian biến số qua từng thế hệ.
- Giúp quan sát rõ tiến trình hội tụ, đa dạng di truyền, và sự thay đổi quần thể theo thời gian.

---

## Ý tưởng thực hiện

- Với bài toán tối ưu hàm $k(x, y, z) = x^2 + y^2 + z^2$, không gian tìm kiếm là 3 chiều.
- Mỗi cá thể là một điểm trong không gian $(x, y, z)$.
- Ở mỗi thế hệ, lấy vị trí giải mã của toàn bộ cá thể, vẽ scatter plot.
- Nếu không thể trực quan 3D trực tiếp, có thể vẽ scatter 2D từng cặp biến: $(x, y)$, $(y, z)$, $(x, z)$ qua các thế hệ.

---

````python
#code vào đây
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

# --- 1. CÁC THAM SỐ CỦA THUẬT TOÁN DI TRUYỀN ---

# Tham số của bài toán
N_VARS = 3  # Số lượng biến (x, y, z)
BOUNDS = [-5, 5]  # Biên giá trị cho mỗi biến

# Tham số của thuật toán
POP_SIZE = 100         # Kích thước quần thể
N_GENERATIONS = 50     # Số thế hệ
BITS_PER_VAR = 16      # Số bit cho mỗi biến (quyết định độ chính xác)
GENOME_LENGTH = N_VARS * BITS_PER_VAR
P_CROSSOVER = 0.8      # Tỷ lệ lai ghép
P_MUTATION = 0.01      # Tỷ lệ đột biến

# --- 2. CÁC HÀM CỐT LÕI CỦA GA ---

# Hàm giải mã: Chuyển chuỗi bit thành các giá trị (x, y, z)
def decode(genome, bounds, bits_per_var, n_vars):
    decoded_values = []
    min_bound, max_bound = bounds
    
    for i in range(n_vars):
        # Tách chuỗi bit cho từng biến
        start = i * bits_per_var
        end = start + bits_per_var
        sub_genome = genome[start:end]
        
        # Chuyển từ nhị phân sang số nguyên
        int_val = int("".join(sub_genome.astype(str)), 2)
        
        # Chuyển về khoảng giá trị [min_bound, max_bound]
        max_int_val = 2**bits_per_var - 1
        value = min_bound + (int_val / max_int_val) * (max_bound - min_bound)
        decoded_values.append(value)
        
    return decoded_values

# Hàm thích nghi (Fitness): Đánh giá mức độ tốt của một cá thể
def fitness_function(individual_coords):
    x, y, z = individual_coords
    k = x**2 + y**2 + z**2
    # Chuyển bài toán min -> max. Thêm 1 để tránh chia cho 0.
    return 1 / (1 + k)

# Hàm chọn lọc (Selection): Dùng phương pháp bánh xe Roulette
def selection(population, fitnesses):
    total_fitness = np.sum(fitnesses)
    probabilities = fitnesses / total_fitness
    # Chọn 2 cá thể cha mẹ
    selected_indices = np.random.choice(len(population), size=2, p=probabilities)
    return population[selected_indices[0]], population[selected_indices[1]]

# Hàm lai ghép (Crossover): Lai ghép 1 điểm cắt
def crossover(parent1, parent2, p_crossover):
    if np.random.rand() < p_crossover:
        # Chọn ngẫu nhiên điểm cắt
        crossover_point = np.random.randint(1, len(parent1) - 1)
        child1 = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
        child2 = np.concatenate((parent2[:crossover_point], parent1[crossover_point:]))
        return child1, child2
    return parent1, parent2

# Hàm đột biến (Mutation): Đảo bit
def mutation(individual, p_mutation):
    for i in range(len(individual)):
        if np.random.rand() < p_mutation:
            individual[i] = 1 - individual[i] # Đảo bit (0 -> 1, 1 -> 0)
    return individual

# --- 3. VÒNG LẶP CHÍNH CỦA THUẬT TOÁN ---

# Khởi tạo quần thể ngẫu nhiên
population = np.random.randint(0, 2, size=(POP_SIZE, GENOME_LENGTH))

# Lưu lại lịch sử tọa độ của quần thể qua các thế hệ để vẽ
population_history = []

print("Bắt đầu quá trình tiến hóa...")
for gen in range(N_GENERATIONS):
    # Giải mã và tính toán độ thích nghi
    decoded_population = np.array([decode(ind, BOUNDS, BITS_PER_VAR, N_VARS) for ind in population])
    fitnesses = np.array([fitness_function(coords) for coords in decoded_population])
    
    # Lưu lại tọa độ của thế hệ hiện tại
    population_history.append(decoded_population)
    
    # In thông tin thế hệ hiện tại
    best_idx = np.argmax(fitnesses)
    best_fitness = fitnesses[best_idx]
    best_solution = decoded_population[best_idx]
    k_value = best_solution[0]**2 + best_solution[1]**2 + best_solution[2]**2
    
    if (gen % 5 == 0 or gen == N_GENERATIONS - 1):
        print(f"Thế hệ {gen}: Best Fitness = {best_fitness:.5f}, "
              f"Best Solution (x,y,z) = ({best_solution[0]:.4f}, {best_solution[1]:.4f}, {best_solution[2]:.4f}), "
              f"k(x,y,z) = {k_value:.6f}")

    # Tạo thế hệ mới
    new_population = []
    for _ in range(POP_SIZE // 2):
        parent1, parent2 = selection(population, fitnesses)
        child1, child2 = crossover(parent1, parent2, P_CROSSOVER)
        child1 = mutation(child1, P_MUTATION)
        child2 = mutation(child2, P_MUTATION)
        new_population.extend([child1, child2])
        
    population = np.array(new_population)

print("Quá trình tiến hóa hoàn tất.")


# --- 4. TRỰC QUAN HÓA BẰNG BIỂU ĐỒ ĐỘNG ---

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Điểm tối ưu (0,0,0) để tham chiếu
ax.scatter([0], [0], [0], c='red', marker='*', s=200, label='Điểm tối ưu (0,0,0)')

# Hàm cập nhật cho mỗi frame của animation
def update(frame):
    # Xóa các điểm của frame trước
    ax.cla()
    
    # Lấy dữ liệu của thế hệ hiện tại (frame)
    current_pop_coords = population_history[frame]
    x_coords = current_pop_coords[:, 0]
    y_coords = current_pop_coords[:, 1]
    z_coords = current_pop_coords[:, 2]
    
    # Vẽ các cá thể của thế hệ hiện tại
    ax.scatter(x_coords, y_coords, z_coords, c='blue', alpha=0.6)
    
    # Giữ lại điểm tối ưu và cài đặt cho trục
    ax.scatter([0], [0], [0], c='red', marker='*', s=200, label='Điểm tối ưu (0,0,0)')
    ax.set_title(f"Phân bố quần thể ở thế hệ {frame}")
    ax.set_xlabel('Biến x')
    ax.set_ylabel('Biến y')
    ax.set_zlabel('Biến z')
    ax.set_xlim(BOUNDS)
    ax.set_ylim(BOUNDS)
    ax.set_zlim(BOUNDS)
    ax.legend()
    ax.grid(True)

# Tạo và hiển thị animation
ani = FuncAnimation(fig, update, frames=N_GENERATIONS, interval=200, repeat=False)
plt.show()
````
